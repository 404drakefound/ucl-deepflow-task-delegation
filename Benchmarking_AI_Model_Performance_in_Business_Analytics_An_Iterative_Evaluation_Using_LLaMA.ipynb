```python
# Cell 1: Imports and Utility Functions
# -------------------------------------------------
# This cell imports all the necessary libraries and defines helper functions.
# These utilities, such as prime_factorization, count_r_in_strawberry, and current_timestamp,
# are used later to simulate iterative logs from LLaMA API calls.
import random
import re
import statistics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Utility: Computes the prime factorization of a given integer.
def prime_factorization(n: int):
    factors = []
    divisor = 2
    temp = n
    while temp >= 2:
        if temp % divisor == 0:
            factors.append(divisor)
            temp //= divisor
        else:
            divisor += 1
    return factors

# Utility: Counts occurrences of the letter "R" (case-insensitive) in "Strawberry".
def count_r_in_strawberry():
    word = "Strawberry"
    return sum(1 for char in word if char.lower() == 'r')

# Utility: Returns a string of the current timestamp.
def current_timestamp():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Run tests for the utility functions.
print("Prime factorization of 9999:", prime_factorization(9999))
print("Number of 'R' in Strawberry:", count_r_in_strawberry())
print("Current timestamp:", current_timestamp())

```

# **2. Simulated Llama 3.1 Evaluation and Data Aggregation**

Here we simulate the iterative evaluations that would be produced by Llama 3.1 (via ollama) for our AI models. We are simulating evaluations for two models (qwen and deepseek) over 10 iterations each and for three prompt types: "Zero-shot," "role-based," and "COT."
For each iteration, we generate a simulated log that contains:


*   A timestamp,

*  Simulated performance metrics (accuracy, average token count, token variance, TPS, inference time),

*   A sample output string,

*   A Boolean flag isCorrect indicating whether the output met a success criterion.


All these simulated outputs are aggregated into a pandas DataFrame following the schema specified in the metaprompt.

**Output Expected:**
You will see printed log messages for each simulated iteration, and then a preview (head) of the aggregated DataFrame showing the first few rows with all columns.

```python
#  Import Libraries and Define Helper Functions
import pandas as pd
import random
from datetime import datetime

def current_timestamp():
    """
    Returns the current time as a formatted string.
    This simulates the timestamp information from a Llama call.
    """
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")
```

```python
# Cell 2: Simulated Llama 3.1 Evaluation and Data Aggregation

# Define experimental parameters
models = ["qwen", "deepseek"]
prompt_types = ["Zero-shot", "role-based", "COT"]
iterations = 10

# Starting prompt_id counter
prompt_id_counter = 100

# List to hold all simulation rows
results = []

def simulate_llama_output(model: str, iteration: int, prompt_type: str):
    """
    Simulates an output log for a given iteration, similar to a Llama 3.1 (ollama) call.

    Returns:
      - A log string with timestamp, accuracy, inference time, and optional convergence message.
      - Simulated average token count.
      - Simulated token variance percentage.
      - Simulated average TPS.
      - Simulated average inference time.
      - A sample output string.
      - A Boolean isCorrect flag.
    """
    timestamp = current_timestamp()
    # Simulate the average token count (e.g., between 100 and 300 tokens)
    avg_token_count = random.randint(100, 300)
    # Simulate token variance as a percentage value
    min_max_variance = round(random.uniform(5, 30), 2)
    # Simulate transactions per second (TPS)
    avg_tps = round(random.uniform(2.0, 5.0), 2)
    # Simulate the average inference time in seconds
    avg_inference_time = round(random.uniform(0.2, 0.8), 3)

    # Simulate accuracy increasing with iterations; deepseek receives a slight bonus.
    accuracy = round(85 + iteration * random.uniform(0.2, 1.0) + (2 if model == "deepseek" else 0), 2)
    # Determine correctness: flag as correct if accuracy is above 90 and a random factor is True.
    is_correct = (accuracy > 90) and random.choice([True, False])
    sample_output = f"Simulated output for {model} at iteration {iteration}"
    # A convergence message is included on the last iteration for demonstration.
    convergence_notice = f"Convergence reached at iteration {iteration}" if iteration == iterations else ""

    output_text = (f"[{timestamp}] Iteration {iteration}: Accuracy: {accuracy:.2f}%, "
                   f"Time: {avg_inference_time:.3f} sec. {convergence_notice}")
    return output_text, avg_token_count, min_max_variance, avg_tps, avg_inference_time, sample_output, is_correct

# Simulate evaluations for every model and prompt type combination.
for model in models:
    for prompt_type in prompt_types:
        for iter_no in range(1, iterations + 1):
            output, avg_token, variance, tps, inf_time, sample_out, is_corr = simulate_llama_output(model, iter_no, prompt_type)
            # Print the simulated log so that users can inspect each iteration's output.
            print(f"Model: {model}, Prompt Type: {prompt_type}, {output}")
            results.append({
                "prompt_id": prompt_id_counter,
                "prompt": f"Sample prompt for {prompt_type}",
                "model_name": model,
                "iteration_no": iter_no,
                "average_token_count": avg_token,
                "min_max_token_variance_%": variance,
                "average_tps": tps,
                "average_inference_time": inf_time,
                "sample_output": sample_out,
                "isCorrect": is_corr,
                "prompt_type": prompt_type
            })
            prompt_id_counter += 1

# Convert the list of results into a DataFrame for further analysis and visualization.
df = pd.DataFrame(results)


```

```python
#  Display Entire Aggregated Data and Prepare CSV Download

# Remove the display limit so that all rows are shown (for our simulated dataset, typically 60 rows)
pd.set_option('display.max_rows', None)

# Print the complete DataFrame in a table format without the index
print("\n--- Aggregated Data ---")
print(df.to_string(index=False))

# Save the entire DataFrame to a CSV file for download
csv_filename = "aggregated_data.csv"
df.to_csv(csv_filename, index=False)
print(f"\nThe aggregated data has been saved to '{csv_filename}'. You can download this file for further analysis.")
```

# **3: Prediction Tables and Correctness Counts**





Here, we create pivot tables for the models qwen and deepseek. Each table shows the results for 10 iterations across the three prompt types (Zero–shot, role-based, COT) by aggregating the isCorrect flag. In addition, we compute the total count of correct predictions (where isCorrect is True) for each model.

```python


# Create prediction pivot tables for 'qwen'
qwen_df = df[df["model_name"]=="qwen"].copy()
qwen_pivot = qwen_df.pivot_table(index="iteration_no", columns="prompt_type",
                                  values="isCorrect", aggfunc=lambda x: ' | '.join(str(int(val)) for val in x))
print("\n--- Prediction Table for qwen ---")
print(qwen_pivot)

# Create prediction pivot tables for 'deepseek'
deepseek_df = df[df["model_name"]=="deepseek"].copy()
deepseek_pivot = deepseek_df.pivot_table(index="iteration_no", columns="prompt_type",
                                         values="isCorrect", aggfunc=lambda x: ' | '.join(str(int(val)) for val in x))
print("\n--- Prediction Table for deepseek ---")
print(deepseek_pivot)

# Count total correct predictions for each model (summing Boolean values)
correct_counts = df.groupby("model_name")["isCorrect"].sum().reset_index()
correct_counts.columns = ["model_name", "total_correct_predictions"]
print("\n--- Total Correct Predictions ---")
print(correct_counts.to_string(index=False))
```
